### Estrutura with


```python
arquivo = open("arquivo.txt", "w")
arquivo.write("Nao sei o que faco")
```

```python
arquivo = open("arquivo.txt", "r", encoding="utf-8")
print(arquivo.read())
arquivo.close()
```

A estrutura "with" minimiza o uso de linhas e deixa o codigo mais limpo, ou seja, ela e mais funcional e tambem de boa pratica utilizala. Ela retira a necessidade de fechar o arquivo, pois faz automaticamente. Ela funciona junto com a funcao "open" onde temos o modo de leitura "r - read" e de escrita "w - write"

```python
with open("arquivo.txt", "r", encoding="utf-8") as arquivo:
    data = arquivo.read()

data
```
